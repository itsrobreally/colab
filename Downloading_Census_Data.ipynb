```python
import pandas as pd
url = "https://svi.cdc.gov/Documents/Data/2020_SVI_Data/CSV/SVI2020_US_COUNTY.csv"
df_svi = pd.read_csv(url)
df_svi.head()
```

```python
import pandas as pd
url = "https://www2.census.gov/geo/docs/reference/codes2020/national_county2020.txt"
df_fips = pd.read_csv(url, sep='|')
df_fips.head()
```

```python
from io import BytesIO
from zipfile import ZipFile
import urllib.request
import pandas as pd
url = "https://prd-tnm.s3.amazonaws.com/StagedProducts/GeographicNames/FederalCodes/FedCodes_National_Text.zip"
with ZipFile(BytesIO(urllib.request.urlopen(url).read())) as my_zip_file:
  for contained_file in my_zip_file.namelist():
    print(contained_file)
  df_bigfips = pd.read_csv(BytesIO(my_zip_file.open("Text/FederalCodes_National.txt").read()), sep='|', low_memory=False)
df_bigfips.head()
```

```python
from io import BytesIO
from zipfile import ZipFile
import urllib.request
import pandas as pd
url = "https://www2.census.gov/programs-surveys/cbp/datasets/2020/cbp20co.zip"
with ZipFile(BytesIO(urllib.request.urlopen(url).read())) as my_zip_file:
  for contained_file in my_zip_file.namelist():
    print(contained_file)
  df_cbp_data = pd.read_csv(BytesIO(my_zip_file.open("cbp20co.txt").read()))
df_cbp_data.head()
```

```python
import pandas as pd
import re
df_cbp_names = pd.read_excel("https://www.census.gov/naics/2022NAICS/2-6%20digit_2022_Codes.xlsx", engine=None)

f = lambda x: re.sub('[\s\.:]+','_', x, flags = re.IGNORECASE)
df_cbp_names.rename(columns = f, inplace = True)

df_cbp_names.head()
```

Let's try to write all of this to a sqlite db

```python
import sqlite3

# Connect to the database (in memory for now)
#conn = sqlite3.connect(':memory:')
conn = sqlite3.connect('/content/census.sqlite3')


# Write the dataframes to tables
conn.execute("DROP TABLE IF EXISTS svi; ")
conn.execute("DROP TABLE IF EXISTS fips; ")
conn.execute("DROP TABLE IF EXISTS bigfips; ")
conn.execute("DROP TABLE IF EXISTS cbp_data; ")
conn.execute("DROP TABLE IF EXISTS cbp_names; ")
conn.commit()

df_svi.to_sql('svi', conn)
df_fips.to_sql('fips', conn)
df_bigfips.to_sql('bigfips', conn)
df_cbp_data.to_sql('cbp_data', conn)
df_cbp_names.to_sql('cbp_names', conn)

# Close the connection
conn.close()
```

```python
conn = sqlite3.connect('/content/census.sqlite3')

#cursor = conn.execute('PRAGMA table_info([cbp_names]);')

cursor = conn.execute(''' SELECT d.location, d.E_TOTPOP population, d.EPL_UNEMP unemployed_pct, d.EP_POV150 below_150_poverty,
  c.`2022_NAICS_US_Title`, b.feature_name, b.state_name, a.emp
FROM cbp_data a
LEFT JOIN bigfips b ON a.fipstate = b.state_numeric AND a.fipscty = b.county_numeric and b.census_class_code like 'H%'
JOIN cbp_names c on round(a.naics) = round(c.`2022_NAICS_US_Code`) and round(a.naics) between 10 and 99 -- 2 digit codes only
join svi d on round(b.state_numeric)*1000 + round(b.county_numeric) = round(d.fips) and b.census_class_code like 'H%'
LIMIT 30
                      ;''')

for row in cursor:
  print(row)
conn.close()
```
